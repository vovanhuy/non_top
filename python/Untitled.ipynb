### Read barcodes from file

```python
import matplotlib.pyplot as plt
%matplotlib inline
def visualize_barcode(path_to_barcode):
    print ("Visualizing barcodes in file " + path_to_barcode)
#     path_to_barcode = "/home/huy/my_document/3x/tda/td5/B.barcode"
    barcode = {}
    # for each dimension, compute max_time and min_time to
    # set x_axis for the figure used to visualize barcodes
    max_time = [0 for i in range(10)]
    min_time = [float("Inf") for i in range(10)]
    with open(path_to_barcode, 'r') as f:
        for line in f:
            try:
                line = line.splitlines()[0].split(" ")
                # get dimension, begin_time and end_time of current barcode
                dim = int(line[0])
                begin_time = float(line[1])
                end_time = float(line[2])
                if begin_time == end_time:
                    continue
                # update min_time and max-time
                min_time[dim] = min(min_time[dim], begin_time)
                if end_time != float("Inf") and end_time > max_time[dim]:
                    max_time[dim] = end_time
                # append current barcode to the list of dim-dimensional barcodes 
                if dim in barcode:
                    barcode[dim].append([begin_time, end_time])
                else:
                    barcode[dim] = [[begin_time, end_time]]
            except Exception:
                continue

    for dim in range(len(max_time)):
        if max_time[dim] == 0:
            max_time[dim] = 10
        max_time[dim] += 1

    for dim in range(len(min_time)):
        if min_time[dim] == -float("Inf"):
            min_time[dim] = 0
    print(max_time)
    print(min_time)
    for dim in barcode.keys():
        print(len(barcode[dim]))
    # plot barcodes
    vertical_length = 2
    for dim in barcode.keys():
        num_of_lines = len(barcode[dim])
        plt.figure()
        for i, bc in enumerate(barcode[dim]):
            if bc[1] - bc[0] < 0.00:
                continue
            if bc[1] != float("Inf"):
                plt.plot(bc, [1.8 - i/num_of_lines, 1.8 - i/num_of_lines], color ='blue')
            else:
                plt.plot([bc[0], max_time[dim]], [1.8 - i/num_of_lines, 1.8 - i/num_of_lines], color='blue')
        plt.axis([min_time[dim], max_time[dim], 0, vertical_length])
        plt.show()
        
```

### Visualize barcodes for d-spheres

```python
for i in range(1, 11):
    path_to_barcode = "/home/huy/my_document/3x/tda/td5/" + str(i) + "_sphere"
    visualize_barcode(path_to_barcode)
```

### Visualize barcodes for d-balls

```python
for i in range(1, 11):
    path_to_barcode = "/home/huy/my_document/3x/tda/td5/" + str(i) + "_ball"
    visualize_barcode(path_to_barcode)
```
