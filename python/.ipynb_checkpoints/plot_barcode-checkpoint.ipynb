### Read barcodes from file

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline


def read_barcode(path_to_barcode, filter_threshold=0, scale=None):
    count = 0
    print("Reading barcodes from files")
    barcode = {}
    # for each dimension, compute max_time and min_time to
    # set x_axis for the figure used to visualize barcodes
    max_time = [-float("Inf") for i in range(10)]
    min_time = [float("Inf") for i in range(10)]
    with open(path_to_barcode, 'r') as f:
        for line in f:
            try:
                line = line.splitlines()[0].split(" ")
                # get dimension, begin_time and end_time of current barcode
                dim = int(line[0])
                begin_time = float(line[1])
                end_time = float(line[2])
                if scale == 'log':
                    if begin_time != 0:
                        begin_time = np.log(begin_time)
                    else:
                        begin_time = -20
                    if end_time != float("Inf"):
                        end_time = np.log(end_time)
                        
                if dim == 1 and (end_time == float("Inf") or end_time - begin_time > 2.5):
                    count += 1
                                
                
                # eliminate barcodes whose length is smaller than filter_threshold
                if end_time != float("Inf") and end_time - begin_time < filter_threshold:
                    continue
                
                if begin_time == end_time:
                    continue
                # update min_time and max-time
                min_time[dim] = min(min_time[dim], begin_time)
                if end_time != float("Inf") and end_time > max_time[dim]:
                    max_time[dim] = end_time
                # append current barcode to the list of dim-dimensional barcodes 
                if dim in barcode:
                    barcode[dim].append([begin_time, end_time])
                else:
                    barcode[dim] = [[begin_time, end_time]]
            except Exception:
                continue

    for dim in range(len(max_time)):
        if max_time[dim] == -float("Inf"):
            max_time[dim] = min_time[dim] + 2
        max_time[dim] += 1
        
    return barcode, max_time, min_time


def visualize_barcode(path_to_barcode, filter_threshold=0, scale=None):
    barcode, max_time, min_time = read_barcode(path_to_barcode, filter_threshold, scale)
    
    # plot barcodes
    vertical_length = 2
    for dim in barcode.keys():
        num_of_lines = len(barcode[dim])
        plt.figure()
        for i, bc in enumerate(barcode[dim]):
            if bc[1] - bc[0] < 0.00:
                continue
            if bc[1] != float("Inf"):
                if bc[0] == bc[1]:
                    plt.plot([bc[0], 1.8 - 1.6*i/num_of_lines], 'ro')
                else:
                    plt.plot(bc, [1.8 - 1.6*i/num_of_lines, 1.8 - 1.6*i/num_of_lines], color ='blue', linewidth=3.0)
            else:
                plt.plot([bc[0], max_time[dim]], [1.8 - 1.6*i/num_of_lines, 1.8 - 1.6*i/num_of_lines], 
                         color='blue', linewidth=3.0)
        plt.ylabel("H_" + str(dim))
        plt.axis([min_time[dim], max_time[dim], 0, vertical_length])
        plt.show()
        
```

### Visualize barcodes for d-spheres

```python
for i in range(1, 11):
    path_to_barcode = "/home/huy/my_document/3x/tda/td5/" + str(i) + "_sphere"
    print("Barcodes of " + str(i) + "-sphere")
    visualize_barcode(path_to_barcode)
```
